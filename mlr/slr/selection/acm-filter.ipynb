```python
import pandas as pd
import bibtexparser
from bibtexparser.bparser import BibTexParser
from bibtexparser.customization import convert_to_unicode

# Função para ler o arquivo .bib
def read_bib_file_from_drive(file_path):
    with open(file_path, 'r', encoding='utf-8') as bibfile:
        parser = BibTexParser()
        parser.customization = convert_to_unicode
        bib_database = bibtexparser.load(bibfile, parser=parser)
    return bib_database.entries

# Função para realizar a pesquisa no arquivo .bib
def search_bib_data(bib_data, search_string):
    matching_entries = []
    for entry in bib_data:
        text_to_search = entry.get('title', '') + ' ' + entry.get('abstract', '') + ' ' + entry.get('keywords', '')
        if all(any(term.lower() in text_to_search.lower() for term in group) for group in search_string):
            matching_entries.append(entry)
    return matching_entries

# Função para criar um DataFrame com os resultados
def create_dataframe(results):
    df = pd.DataFrame(results)
    return df

# Função para salvar o DataFrame em um arquivo CSV
def save_to_csv(df, output_path):
    df.to_csv(output_path, index=False)
    print(f'DataFrame salvo em {output_path}')

# Caminho para o arquivo .bib
bib_file_path = 'drive/MyDrive/Mestrado/Revisão Multivocal/versão 5x/acm.bib'

# String de busca
search_string = [
    ["software ecosystems", "developer community", "open source community"],
    ["life cycle", "maintenance", "updates", "enhancement", "evolution", "vital signs", "health", "healthy"],
    ["death", "dead", "die", "dying", "static", "sick", "discontinu", "unmaintained", "End-of-life", "demise", "aging"]
]

# Lendo o arquivo .bib
bib_data = read_bib_file_from_drive(bib_file_path)

# Realizando a pesquisa e criando o DataFrame
results = search_bib_data(bib_data, search_string)
df_results = create_dataframe(results)

# Caminho para salvar o arquivo CSV
output_csv_path = 'drive/MyDrive/Mestrado/Revisão Multivocal/versão 5x/acm.csv'

# Salvando o DataFrame em um arquivo CSV
save_to_csv(df_results, output_csv_path)
```
